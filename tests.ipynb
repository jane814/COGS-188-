```python
from engine import Game, Player
# setting up pygame
import pygame 
import matplotlib.pyplot as plt
import numpy as np

pygame.init()
pygame.font.init()
pygame.display.set_caption('Battleship')
font = pygame.font.SysFont("fresansttf", 100)

num_games = 500
n_shots = []
mean_shots = []
wins1 = 0
wins2 = 0
grid_sizes = [10 , 15, 20 , 25]
for grid_size in grid_sizes:
    for i in range(num_games):
        game = Game(human1=False, human2=False, grid_size=grid_size)
        while not game.over:
            if game.player1_turn:
                game.q_learning_policy()
            else:
                game.q_learning_policy()
        n_shots.append(game.n_shots)
    mean_shots.append(np.mean(n_shots))

plt.plot(grid_sizes, mean_shots ) 
plt.xlabel('Grid Sizes')
plt.ylabel('Average Number of Shots')
plt.title('Average Number of Shots vs Grid Size')

plt.show()
```
